# Ablation Study: Variance Decomposition

```python
from configparser import ConfigParser
from pathlib import Path
import pandas as pd

from current_research_feature_effects.plotting.plots import plot_variance_table
from current_research_feature_effects.plotting.table_formatting import (
    get_concatenated_results,
    get_grouped_df,
    filter_variance_metrics,
    flatten_variance_df,
)
```

```python
EXPERIMENT_PATH = Path("../experiments/ablation_study_variance_decomposition_new")
config = ConfigParser()
config.read(EXPERIMENT_PATH / "ablation_study_variance_decomposition.ini")
```

```python
MAIN_EXPERIMENT_PATH = Path("../experiments/main_study_parallel")
main_config = ConfigParser()
main_config.read(MAIN_EXPERIMENT_PATH / "main_study.ini")
```

## SimpleNormalCorrelated

### PDP

```python
df_pdp_snc = get_concatenated_results(
    "SimpleNormalCorrelated",
    "pdp",
    config,
    main_config,
    main_experiment_path=MAIN_EXPERIMENT_PATH,
    experiment_path=EXPERIMENT_PATH,
)
df_pdp_snc = get_grouped_df(df_pdp_snc)
df_pdp_snc = filter_variance_metrics(df_pdp_snc)
df_pdp_snc
```

```python
df_pdp_snc_var = flatten_variance_df(df_pdp_snc)
plot_variance_table(
    df_pdp_snc_var,
    models=["XGBoost_OT", "XGBoost_OF"],
    type="pdp",
    # save_figs=Path("../paper/img/SNC-all/"),
    # large_font=True,
    # show_title=False,
)
```

## Feynman I.29.16

### PDP

```python
df_pdp_feynman = get_concatenated_results(
    "FeynmanI.29.16",
    "pdp",
    config,
    main_config,
    main_experiment_path=MAIN_EXPERIMENT_PATH,
    experiment_path=EXPERIMENT_PATH,
)
df_pdp_feynman = get_grouped_df(df_pdp_feynman)
df_pdp_feynman = filter_variance_metrics(df_pdp_feynman)
new_order = ['x_1', 'x_2', 'theta_1', 'theta_2', 'd_1', 'd_2']
df_pdp_feynman = df_pdp_feynman.reindex(columns=pd.MultiIndex.from_product([new_order, ['Variance', 'Model Variance', 'MC Variance']]))
df_pdp_feynman
```

```python
df_pdp_feynman_var = flatten_variance_df(df_pdp_feynman)
feature_order = ["x_1", "x_2", "theta_1", "theta_2", "d_1", "d_2"]
df_pdp_feynman_var["feature"] = pd.Categorical(df_pdp_feynman_var["feature"], categories=feature_order, ordered=True)
plot_variance_table(
    df_pdp_feynman_var,
    models=["XGBoost_OT", "XGBoost_OF"],
    type="pdp",
    # large_font=True,
    # save_figs=Path("../paper/img/FeynmanI.29.16-all/"),
    # show_title=False,
)
```

### ALE

```python
df_ale_feynman = get_concatenated_results(
    "FeynmanI.29.16",
    "ale",
    config,
    main_config,
    main_experiment_path=MAIN_EXPERIMENT_PATH,
    experiment_path=EXPERIMENT_PATH,
)
df_ale_feynman = get_grouped_df(df_ale_feynman)
df_ale_feynman = filter_variance_metrics(df_ale_feynman)
df_ale_feynman = df_ale_feynman.reindex(
    columns=pd.MultiIndex.from_product([new_order, ["Variance", "Model Variance", "MC Variance"]])
)
df_ale_feynman
```

```python
df_ale_feynman_var = flatten_variance_df(df_ale_feynman)
plot_variance_table(
    df_ale_feynman_var,
    models=["XGBoost_OT", "XGBoost_OF"],
    type="ale",
    large_font=True,
    save_figs=Path("../paper/img/FeynmanI.29.16-all/"),
    show_title=False,
)
```
